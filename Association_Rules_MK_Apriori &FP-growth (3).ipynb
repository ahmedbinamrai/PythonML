```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('MKSFR-1-9-2024_30-9-2024.csv', sep=',')
```

```python
df = pd.DataFrame(dataset)
```

```python
df
```

```python
df1 = df.groupby("ReceiptNumber")["ItemId"].agg(list).reset_index().fillna(0).set_index("ReceiptNumber")
df1
```

        Mlxtend section

```python
!pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_ary = te.fit(df1["ItemId"]).transform(df1["ItemId"])
df2 = pd.DataFrame(te_ary, columns=te.columns_)
df2
```

        Apriori section

```python
from mlxtend.frequent_patterns import apriori 
frequent_itemsets = apriori(df2, min_support= 0.02, use_colnames=True)
```

```python
frequent_itemsets
```

```python
frequent_itemsets.to_csv('result-freq-9-2024(3).csv', index=False)
```

```python
from mlxtend.frequent_patterns import association_rules
result = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.02)
result[['antecedents','consequents','support','confidence','lift']]
```
